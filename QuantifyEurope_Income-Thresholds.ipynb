```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
```

Data source:
Mean and median income by age and sex - EU-SILC survey [ilc_di03]

Last update: 30-05-2017

http://ec.europa.eu/eurostat/en/web/products-datasets/-/ILC_DI03

Sex: Total

Age: Total

Indicator: Median equivalized income in Euro

Year: 2015 (2016 not yet available for many countries)

```python
df = pd.read_csv("ilc_di03_1_Data-2017-06-14.csv")
```

```python
# select age total
df = df[df["AGE"]=="Total"]
```

```python
# select relevant columns
df = df[["GEO", "Value"]]
```

```python
def float_conversion(value_str):
    try:
        value=float(value_str.replace(",",""))
        return value
    except:
        pass
```

```python
# convert values to actual float
df["Median_income"]=df["Value"].apply(float_conversion)
```

```python
# select relevant columns
df=df[["GEO", "Median_income"]]
```

```python
# calculate the "at risk of poverty" threshold
df["poverty_theshold"]= df["Median_income"]*0.6
```

```python
# calculate the middle class threshold
df["middle_class_threshold"]= df["Median_income"] *0.7
```

```python
# calculate the upper class threshold
df["upper_class_threshold"]=df["Median_income"]*1.5
```

```python
# inspect values
df.describe()
```

```python
df.sort_values(by="upper_class_threshold")
```

```python
#save sorted table into csv for further d3 work
df.sort_values(by="upper_class_threshold").to_csv("income_threshold_sorted.csv")
```
